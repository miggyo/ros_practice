```python
import numpy as np
import matplotlib.pyplot as plt 

class Drawsin:
    def __init__(self, amp,freq,bias,end_time) : 
        self.amp = amp
        self.freq = freq
        self.bias = bias
        self.end_time = end_time

    def calc_sin(self):
        self.t = np.arange(0, self.end_time, 0.01)
        return self.amp *np.sin(2*np.pi*self.freq*self.t) + self.bias
    
    def draw_sin(self):
        y = self.calc_sin()
        plt.figure(figsize=(12,6))
        plt.plot(self.t, y)
        plt.grid()
        plt.show()
```

```python
ds = Drawsin(1,1,0,3)
ds.draw_sin()
```

```python
ds.freq = 2
ds.draw_sin()
```
