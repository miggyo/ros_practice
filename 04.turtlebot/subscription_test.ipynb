```python
import rclpy as rp
from turtlesim.msg import Pose
```

```python
rp.init()
test_node = rp.create_node('sub_test')
```

```python
def callback(data):
    print("------->")
    print("/turtle1/pose : ",data)
    print("X: ",data.x)
    print("Y: ",data.y)
    print("Theta: ",data.theta)
    
```

```python
test_node.create_subscription(Pose,"/turtle1/pose",callback,10 )
```

```python
rp.spin(test_node)
```

```python
import rclpy as rp
from turtlesim.msg import Pose
rp.init()
test_node = rp.create_node('sub_test')
cnt = 0
def callback(data):
    global cnt
    cnt += 1
    print(">",cnt, " ->X : ",data.x, ",Y : ", data.y)
    if cnt > 3:
        raise Exception("subscription Stop")
```

```python
test_node.create_subscription(Pose,"/turtle1/pose",callback,10 )
```

```python
rp.spin(test_node)
```

```python
import rclpy as rp
from geometry_msgs.msg import Twist

test_node = rp.create_node('pub_test')
```

```python
msg = Twist()
print(msg)
```

```python
msg.linear.x = 2.0
print(msg)
```

```python
pub = test_node.create_publisher(Twist, '/turtle1/cmd_vel',10)
pub.publish(msg)
```

```python
cnt = 0
def timer_callback():
    global cnt

    cnt += 1

    print(cnt)
    pub.publish(msg)

    if cnt > 3:
        raise Exception("publisher stop")
```

```python
timer_period  = 0.1
timer = test_node.create_timer(timer_period, timer_callback)
rp.spin(test_node)
```

```python
test_node.destroy_node()
```
